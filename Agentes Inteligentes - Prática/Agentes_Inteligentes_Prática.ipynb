```python
import os
import pyttsx3 as p
from datetime import datetime
#speak() converts text to speech
p.speak("Oi, eu sou a Sherelayne")
print("Sherelayne: Oi, eu sou a Sherelayne")
#loops until user specify to quit
while True: 
    userinput = input()
    
    #Prints 
    if("hi" in userinput) or ("oi" in userinput) or ("olá" in userinput):
        p.speak("E aí?")
        print("Sherelayne: E aí?")
    elif("como vai?" in userinput):
        p.speak("Estou bem, e você?")
        print("Sherelayne: Estou bem, e você?")
    elif("bem" in userinput):
        p.speak("Que bom")
        print("Sherelayne: Que bom")
    elif("quem é a sua família?" in userinput):
        p.speak("Meu pai é o Xerox, minhas tias são Autenticada e Fotocópia, minhas irmãs são Sherlayne e Shekira e meus irmãos são Brucesfield, Carlos Eduardo e Carimbo")
        print("Sherelayne: Meu pai é o Xerox, minhas tias são Autenticada e Fotocópia, minhas irmãs são Sherlayne e Shekira e meus irmãos são Brucesfield, Carlos Eduardo e Carimbo")
    elif("que horas são?" in userinput):
        now = datetime.now()
        current_time = now.strftime("%H:%M:%S")
        p.speak("Agora são %s"%current_time)
        print("Sherelayne: Agora são %s"%current_time)
    elif("que dia é hoje?" in userinput):
        now = datetime.now()
        current_time = now.strftime("dia %d do mês %m do ano %Y")
        p.speak("Hoje é %s"%current_time)
        print("Sherelayne: Hoje é %s"%current_time)
    #It stops the execution
    elif ("exit" in userinput) or ("quit" in userinput):
        p.speak("Foi um prazer falar com você")
        print("Sherelayne: Foi um prazer falar com você!")
        break
    #If user inputs anything wrong, it shows don't support
    else:
        print("Não entendi")
```
